# Задача: Анализ эффективности нового алгоритма рекомендаций в условиях асимметричного и зашумленного распределения
### Контекст:
Крупная стриминговая платформа (аналог YouTube или Twitch) разработала новый алгоритм рекомендаций контента (алгоритм B), который, как ожидается, должен увеличить среднее время просмотра на пользователя по сравнению со старым алгоритмом (A).

Было проведено A/B-тестирование:

- **Группа A (Контрольная)**: 5000 пользователей, старый алгоритм.

- **Группа B (Тестовая)**: 5000 пользователей, новый алгоритм.

По итогам двухнедельного теста вы собрали данные по ключевой метрике — **«Общее время просмотра на пользователя за тест (в минутах)»**.

**Данные и их особенности:**

**Не нормальность:** Распределение метрики в обеих группах сильно отличается от нормального. Оно имеет типичный для поведенческих данных "длинный хвост":

- Большинство пользователей смотрят немного (пик в районе 0-50 минут).

- Небольшая группа "супер-зрителей" смотрит очень много (сотни и тысячи минут), формируя длинный правый хвост распределения.

**Асимметрия влияния:** Новый алгоритм B целенаправленно меняет опыт именно активных пользователей, в то время как для "пассивных" зрителей разница неочевидна. Это означает, что различие между группами может быть не сдвигом всего распределения, а изменением в правом хвосте.

**Наличие выбросов:** В данных присутствуют экстремальные выбросы (например, пользователи, которые оставляли видео фоном на всю ночь).

Вам предоставлен датасет _ab_test_results.csv_ со следующими колонками:

 - _user_id_

 - _group (A или B)_

 - _total_watch_time_min_

### Задания:


- **Проведите тест Манна-Уитни (Mann-Whitney U test)**. Сформулируйте нулевую и альтернативную гипотезы. Интерпретируйте p-value. Можно ли на его основе сделать однозначный вывод о превосходстве алгоритма B?

- Чтобы исследовать "асимметричное влияние", проведите анализ разниц не только в центре распределения, но и в хвостах. **Рассчитайте и визуализируйте разницу между группами для 75-го, 90-го и 95-го процентилей**.

- **Сформулируйте итоговый вывод:**

    - Эффективен ли новый алгоритм B?

    - На какую именно группу пользователей он оказывает наибольшее влияние?

    - Какой метод анализа оказался наиболее информативным в данном случае и почему?

    - Дайте рекомендацию о возможности запуска алгоритма на всю аудиторию.



```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Load CSV
df = pd.read_csv('ab_test_results.csv')

group_a_data = df[df['group'] == 'A']['total_watch_time_min']
group_b_data = df[df['group'] == 'B']['total_watch_time_min']

# Display summary statistics
print("Dataset Summary:")
print(f"Total users: {len(df)}")
print(f"Group A: {len(df[df['group'] == 'A'])} users")
print(f"Group B: {len(df[df['group'] == 'B'])} users")
print("\nDescriptive Statistics by Group:")
print(df.groupby('group')['total_watch_time_min'].describe())

print("\nKey Percentiles by Group:")
percentiles = [0.75, 0.9, 0.95]
for p in percentiles:
    p_val = p * 100
    a_val = np.percentile(group_a_data, p_val)
    b_val = np.percentile(group_b_data, p_val)
    diff = b_val - a_val
    print(f"{p_val:.0f}th percentile: A={a_val:.1f}, B={b_val:.1f}, Difference={diff:.1f}")

# Test the differences
from scipy import stats
mw_stat, mw_p = stats.mannwhitneyu(
    group_a_data,
    group_b_data,
    alternative= 'two-sided' #'less' #'greater' #'two-sided'
)

print(f"\nMann-Whitney U Test: p-value = {mw_p:.6f}")
```

## Тест Манна-Уитни (Mann-Whitney U test)

### Формулировка гипотез:

**Нулевая гепотеза ($ H_0 $):**

Распределение метрики `total_watch_time_min` в группе **B** (новый алгоритм) не превышает распределение в группе **A** (старый алгоритм).
Или: медианное время просмотра в группе **B** <= медианному времени в группе **A**.

**Альтернативная гепотеза ($ H_1 $):**

Распределение метрики `total_watch_time_min` в группе **B** стохастически больше (сдвинуто в сторону больших значений), чем в группе **A**. То есть новый алгоритм эффективнее.


## Проведение теста с правильной альтернативой

Для проверки превосходства алгоритма **B** используем **односторонний** тест с альтернативной гипотезой `greater` (B > A).

```python
mw_stat, mw_p_one_sided = stats.mannwhitneyu(
    group_b_data,  # Помещаем группу B на первое место для логики "greater"
    group_a_data,
    alternative='greater'
)
```

```python
print(f"\nРезультаты теста:")
print(f"• Статистика U = {mw_stat:.0f}")
print(f"• p-value (односторонний, B > A) = {mw_p_one_sided:.6f}")
print(f"• p-value (двусторонний) = {mw_p_two_sided:.6f}")
print()
```

### Интерпретация p-value

```python
alpha = 0.05

print(f"Уровень значимости α = {alpha}")
print(f"Полученное p-value = {mw_p_one_sided:.6f}")
```

```python
if mw_p_one_sided < alpha:
    print(f"\np-value ({mw_p_one_sided:.6f}) < α ({alpha}): ОТВЕРГАЕМ нулевую гипотезу.")
    print("  Существуют статистически значимые свидетельства того, что новый алгоритм B увеличивает время просмотра пользователей.")
else:
    print(f"\np-value ({mw_p_one_sided:.6f}) ≥ α ({alpha}): НЕ ОТВЕРГАЕМ нулевую гипотезу.")
    print("  Нет достаточных оснований утверждать, что алгоритм B лучше.")
print()
```

### Ответ на вопрос об однозначности вывода

Можно ли на основе этого теста сделать однозначный вывод о превосходстве алгоритма B?

**Ответ: нет**, по следующим причинам:

1. **Тест Манна-Уитни проверяет общий сдвиг распределений, но не учитывает характер этого сдвига.**  
   Он отвечает на вопрос "различаются ли распределения?", но не показывает, как именно они различаются.

2. **Как видно из анализа процентилей, эффект неравномерен:**  
   • Для 75-го процентиля: разница всего **+1.4 минуты**  
   • Для 95-го процентиля: разница уже **+34.3 минуты**  
   Это означает, что алгоритм работает по-разному для разных сегментов пользователей.

3. **Тест "видит" только средний ранг, но не показывает:**  
   • На каких именно пользователях концентрируется эффект  
   • Является ли разница клинически/бизнесово значимой  
   • Не вызван ли результат единичными выбросами

4. В контексте задачи с асимметричным распределением и целевым воздействием на активных пользователей, тест Манна-Уитни дает лишь **предварительный вывод**. Основной вывод должен основываться на анализе процентилей.

**Вывод:**

Хотя тест показывает статистически значимое различие (p = 0.0227), он **не даёт однозначного ответа о превосходстве алгоритма B** без дополнительного анализа распределения эффекта. Для полноценной оценки необходимо изучать не только общий сдвиг, но и то, как эффект распределяется по разным сегментам пользователей.

## Исследование асимметричного влияния

### Расчет процентилей

```python
percentiles = [75, 90, 95]

print()
print("Процентиль | Группа A | Группа B | Разница (B-A)")
print("-" * 48)

for p in percentiles:
    a_val = np.percentile(group_a_data, p)
    b_val = np.percentile(group_b_data, p)
    diff = b_val - a_val
    print(f"{p:3d}-й      | {a_val:7.1f} | {b_val:7.1f}   | {diff:+.1f} мин")
print()
```

### Визуализация разницы процентилей

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4), dpi=300)

x_pos = np.arange(len(percentiles))
diff_values = [np.percentile(group_b_data, p) - np.percentile(group_a_data, p) for p in percentiles]

bars = ax1.bar(x_pos, diff_values, color=['lightblue', 'lightgreen', 'salmon'])
ax1.set_xlabel('Процентиль')
ax1.set_ylabel('Разница (B - A), мин')
ax1.set_title('Разница времени просмотра по процентилям')
ax1.set_xticks(x_pos)
ax1.set_xticklabels([f'{p}%' for p in percentiles])
ax1.axhline(y=0, color='black', linestyle='-', linewidth=0.5)

# Добавляем значения на столбцы
for bar, val in zip(bars, diff_values):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
             f'{val:+.1f}', ha='center', va='bottom' if val > 0 else 'top')

width = 0.35
x = np.arange(len(percentiles))

ax2.bar(x - width/2, [np.percentile(group_a_data, p) for p in percentiles], 
        width, label='Группа A', alpha=0.8, color='skyblue')
ax2.bar(x + width/2, [np.percentile(group_b_data, p) for p in percentiles], 
        width, label='Группа B', alpha=0.8, color='lightcoral')

ax2.set_xlabel('Процентиль')
ax2.set_ylabel('Время просмотра, мин')
ax2.set_title('Сравнение процентилей между группами')
ax2.set_xticks(x)
ax2.set_xticklabels([f'{p}%' for p in percentiles])
ax2.legend()

plt.tight_layout()
plt.show()
```

### Вывод

Новый алгоритм B оказывает асимметричное влияние:

- **75-й процентиль:** незначительный эффект (+1.4 мин)  
- **90-й процентиль:** умеренный эффект (+4.7 мин)  
- **95-й процентиль:** сильный эффект (+34.3 мин)  

Алгоритм **наиболее эффективен** для самых активных пользователей (верхние 5-10% по времени просмотра).

### Итоговый вывод

**1. Эффективен ли алгоритм B?**  
Да, но работает избирательно. В среднем время просмотра выросло с 56 до 71 минуты, но это "среднее" обманчиво.

**2. Для кого алгоритм работает лучше всего?**

Алгоритм "раскачивает" уже вовлеченных, а не привлекает новых.

**3. Какой метод анализа дал понимание?**  
Сравнение процентилей, а не средних или тестов.  
Только так стало видно, что эффект растёт к хвосту распределения.  
График разницы по процентилям нагляднее таблиц со статистикой.

**4. Запускать ли на всех?**  
Да, запускать, но следить.  
- Плюс: активные зрители смотрят ещё больше — это ценно для платформы  
- Минус: пассивных пользователей алгоритм почти не затронул  
- Риск: неясно, не надоест ли топовым зрителям контент быстрее  

Алгоритм хорош для "раскачки" лояльной аудитории, но не для широкого вовлечения. Запускаем, но смотрим, как поведёт себя через месяц.
